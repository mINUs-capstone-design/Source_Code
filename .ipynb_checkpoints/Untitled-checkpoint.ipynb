```python
import numpy as np
import matplotlib.pyplot as plt

def preemphasis_filter(signal, alpha=0.97):
    """
    Apply pre-emphasis filtering to the input signal.
    
    Parameters:
        signal (ndarray): Input signal.
        alpha (float): Pre-emphasis coefficient (default is 0.97).
    
    Returns:
        ndarray: Filtered signal.
    """
    # Apply pre-emphasis filtering
    filtered_signal = np.append(signal[0], signal[1:] - alpha * signal[:-1])
    return filtered_signal

def plot_spectrum(signal, filtered_signal, sr):
    """
    Plot frequency spectrum of the original and filtered signals.
    
    Parameters:
        signal (ndarray): Original signal.
        filtered_signal (ndarray): Filtered signal.
        sr (int): Sampling rate.
    """
    # Compute frequency spectrum
    spectrum = np.abs(np.fft.fft(signal))
    filtered_spectrum = np.abs(np.fft.fft(filtered_signal))
    
    # Frequency axis
    freq = np.fft.fftfreq(len(signal), d=1/sr)
    
    # Plot frequency spectrum
    plt.figure(figsize=(10, 5))
    plt.plot(freq[:len(freq)//2], spectrum[:len(freq)//2], label='Original Signal')
    plt.plot(freq[:len(freq)//2], filtered_spectrum[:len(freq)//2], label='Filtered Signal')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.title('Frequency Spectrum')
    plt.legend()
    plt.grid(True)
    plt.show()

# 입력 신호
signal = np.array([0.1, 0.3, 0.5, 0.7, 0.9])

# 프리-에멧티브 필터링 적용
filtered_signal = preemphasis_filter(signal)

# 주파수 스펙트럼 비교
plot_spectrum(signal, filtered_signal, sr=44100)
```

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

# 이동평균 필터링 함수
def moving_average_filter(data, window_size):
    kernel = np.ones(window_size) / window_size
    return np.convolve(data, kernel, mode='same')

# 음성 파일 로드
audio_path = 'your_audio_file.wav'
y, sr = librosa.load(audio_path)

# 멜 스펙트로그램 생성
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)

# dB로 변환
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

# 필터링 적용
window_size = 10
filtered_mel_spectrogram_db = moving_average_filter(mel_spectrogram_db, window_size)

# 필터링 전후 멜 스펙트로그램 시각화
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
librosa.display.specshow(mel_spectrogram_db, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Original Mel Spectrogram')

plt.subplot(2, 1, 2)
librosa.display.specshow(filtered_mel_spectrogram_db, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Filtered Mel Spectrogram')

plt.tight_layout()
plt.show()
```
